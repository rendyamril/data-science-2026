```python
#Q1: Muat Dataset Titanic

import pandas as pd
import numpy as np

URL = "https://raw.github.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
print(df.shape, df.columns.tolist())

#A1: dataset Titanic memiliki 891 baris dan 12 kolom.
```

```python
#Q2: Kolom mana yang memiliki nilai kosong (missing) ?

print(df.isnull().sum())

#A2: ada 3 kolom yang memiliki nilai missing, yaitu age, cabin dan embarked.
```

```python
#Q3: Berapa persen penumpang yang selamat (Survived=1) ?

selamat = df[df['Survived']==1]
survival_rate = len(selamat) / len(df)
({
    "Total Penumpang": [len(df)],
    "Jumlah Selamat": [df['Survived'].sum()],
    "Persentase Selamat (%)": [survival_rate * 100]
})

#A3: Penumpang yang berhasil selamat ialah 38.4% (38.38383838383838).
```

```python
#Q4: Rata-rata usia penumpang wanita kelas 1 yang selamat?

wanita_kelas1_selamat = df[
    (df['Sex'] == 'female') &
    (df['Pclass'] == 1) &
    (df['Survived'] == 1)
]
({
    "Jumlah Penumpang": [len(wanita_kelas1_selamat)],
    "Rata-rata Usia": [wanita_kelas1_selamat['Age'].mean()]
})

#A4: Rata-rata usia penumpang wanita kelas 1 yang selamat ialah 35 tahun.
```

```python
#Q5: Kelas penumpang mana yang tingkat keselamatan tertinggi ?

survival_per_kelas = df.groupby('Pclass')['Survived'].mean()

hasil = survival_per_kelas.reset_index()
hasil.columns = ["Kelas", "Rata-rata"]
print(hasil)

#A5: Tingkat keselamatan tertinggi pada kelas penumpang ialah kelas 1.
```
